```python
#next we try to clean the data by filling out some missing values and dropping records with outliers 
df.isna().sum() # prints out the total number of missing values per column
```

```python
# one way of dealing with null values is filling them with values that make business sense such measures of central tendancy 
df['SMS Service'].fillna("Active", inplace=True)
df['Churn Status'].fillna("Active", inplace=True)
df.isna().sum()
```

```python
# lets now identify outliers using the statistical Z score 
#python provides Scipy library to perform all statistical computations 
from scipy import stats 
import numpy as np 
```

```python
#next we need to Calculate absolute Z-score of a particular column(s) based on the requirement.
#Usually threshold value for Z-score is taken as 3 at higher level and -3 at the lowest. Calculated scores out of this range are considered as outliers.
zscore_PN =np.abs(stats.zscore(df['Phone Number']))
print(np.where(zscore_PN > 3))
```

```python
#from the above output, the array shows the row numbers of rows with outliers 
#lets print any row with the above provided row number starting with row index [3]
print(df.iloc[3])
```

```python
#From the output we can see cleary that phonenumer value of 10 is out of place. 
#thus thess row is completely useless and needs to be dropped 
df=df.drop([ 3,  5,  6, 13, 16])
df.shape
```

```python
# from the above ouptut it is visible that five rows have been dropped from the previous 505 to current 500 rows 
```

```python
#lets save the data to a csv 
df.to_csv("SIT303preprocessedSubscriberdata.csv", encoding='utf-8', index=False)
```

# Data Transformation.
using the Activity.csv dataset
This may involve activities such as 
1. Providing the right schema
2. splitting the fields and creating new colunms to make data more readable
3. Filling the NAN Not Number values which is a pecial floating point used to represent missing values in python 


```python
#Proving the the right schema 
# lets see how we can do this using the activity dataset 
df2=pd.read_csv('ActivityData.csv')
df2.head()
```

```python
#columns for the dataset seems not be defined lets define them based on our understading of what they hold 
df2.columns=('PhoneNumber-SubscriberIdentityNumber', 'Duration', 'Activity','ErrorCode')
```

```python
df2.head()
```

```python
# Now let's split the Phone Number-SubscriberIdentityNumber column
df2[['Activity', 'ErroCode']] = df2['Activity'].str.split('-', n=1, expand=True)
df2.head(10)
```

```python

# Now let's split the Phone Number-SubscriberIdentityNumber column
df2[['PhoneNumber', 'Subscriber Identity Number']] = df2['PhoneNumber-SubscriberIdentityNumber'].str.split('-', n=1, expand=True)
df2.head()
```

```python
#now as we can see the split has happend but the old combined columns still exist so drop them 
df2=df2.drop('PhoneNumber-SubscriberIdentityNumber', axis=1)
df2.head()
```

```python
#now as we can see the split has happend but the old combined columns still exist so drop them 
df2=df2.drop('ErrorCode', axis=1)
df2.head()
```

```python
#As you can see the columns are not in the right order lets reorder them 
df2=df2[['Subscriber Identity Number', 'PhoneNumber', 'Duration', 'Activity', 'ErroCode']]
df2.head(10)
```

```python
#the ErroCode column contains some empty values lets convert them to NAN
df2=df2.replace('', np.NaN)
df2.head(10)
```

```python
df2.isna().sum()
```

```python
#fill the null values with zero
df2['ErroCode'].fillna(0, inplace=True)
df2.head(10)
```

```python
df2.to_csv("SIT303PreprocessedActivitiesData.csv", encoding='utf-8', index=False)
```
